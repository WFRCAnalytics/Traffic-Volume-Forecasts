```python
import pandas as pd
```

This code takes the json files and regenerates CSVs, while grabbing F_AREA and PLANAREA from segments.json

It is to help seed the data so the 'lock' functionality works.

```python
dfSegsWithData = pd.read_json('../_site-art-of-forecasting/data/segments.json')
dfSegsWithData
```

```python
# convert csvs to jsons
#lstSegidFiles = ['model-forecasts', 'aadt', 'linear-forecasts', 'segments']
lstSegidFiles = ['model-forecasts']

for file in lstSegidFiles:
    # Read CSV file
    print(file)

    _df = pd.read_json('../_site-art-of-forecasting/data/' + file + '.json')

    columns_to_drop = ['PLANAREA', 'SUBAREAID', 'CO_NAME', 'F_AREA']

    # Use list comprehension to filter columns that exist in the DataFrame
    columns_to_drop = [col for col in columns_to_drop if col in _df.columns]

    # Now, safely drop these columns if they exist
    _df.drop(columns=columns_to_drop, inplace=True)

    _dfWithSegData = pd.merge(dfSegsWithData, _df, on='SEGID', how='left')

    _dfWithSegData.to_csv('../intermediate/' + file + '.csv') 
    
```
